```python
def clean_ticket_price(value):
    if pd.isna(value):
        return np.nan
    value_str = str(value)
    # Fix bad encoding and remove symbols
    value_str = value_str.replace("â‚¹", "").replace("₹", "").replace(",", "").replace("Cr", "").strip()
    try:
        return float(value_str)
    except ValueError:
        return np.nan

df["Ticket_Price_Cr"] = df["Ticket_Price_Cr"].apply(clean_ticket_price)
```
